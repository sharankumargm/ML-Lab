```python
# 1)Write a NumPy program to multiply two given arrays of same size element-by-
# element.
# First here we Import NumPy library with 'np'
import numpy as np

# Create two NumPy arrays 'nums1' and 'nums2' contain matrices of same size
nums1 = np.array([[2, 5, 2],
              [1, 5, 5]])

nums2 = np.array([[5, 3, 4],
              [3, 2, 5]])

# Here we are printing the contents of 'nums1' two arrays
print("Array1:")
print(nums1)

print("Array2:")
print(nums2)

# now we perfom element-wise multiplication of arrays using np.multiply()
# np.multiply operation multiplies corresponding elements of the two arrays
print("\nmultiplication of same size element-by-element gives:")
print(np.multiply(nums1, nums2))
```

```python
# 2)Write a NumPy program to find the missing data in a given array.
# First here we Import NumPy library with 'np'
import numpy as np

# Creating a NumPy array 'nums' with provided values, including NaN (Not a Number)
nums = np.array([[3, 2, np.nan, 1],
                 [10, 12, 10, 9],
                 [5, np.nan, 1, np.nan]])

# To print original array
print("Original array:")
print(nums)

# To indicate missing data we use (NaN) in the array using np.isnan() function
# Using the function we get output and in output True represents NaN Values
print("\nThe he missing data of the given array:")
print(np.isnan(nums))
```

```python
# 3)Write a NumPy program to test whether each element of a 1-D array is also present in
# a second array.
# Import the NumPy library with 'np'
import numpy as np

# Creating 'array1' using NumPy
array1 = np.array([0, 10, 20, 40, 60])

# Printing 'array1'
print("Array1: ", array1)

# Creating 'array2'
array2 = np.array([0,40])

# Printing 'array2'
print("Array2: ", array2)

# Comparing 'array1' with 'array2'
print("Compare each element of array1 and array2")
print(np.in1d(array1, array2))

```

```python
# 4)Write a NumPy program to save a NumPy array to a text file.
import numpy as np

# Create a sample NumPy array
arr = np.array([[1, 2, 3], [4, 5, 6]])

# Save the array to a text file named "array.txt"
np.savetxt("array.txt", arr, delimiter=",")

print("Array saved to text file successfully!")


```
